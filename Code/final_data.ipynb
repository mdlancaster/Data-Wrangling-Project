```python
# import required packages
import pandas as pd
import numpy as np
```

```python
# load the two separate datasets
# nasdaq_final.csv
nasdaq_final = pd.read_csv('nasdaq_final.csv', encoding = 'utf-8')

# fin_stmts.csv
fin_stmts = pd.read_csv('fin_stmts.csv', encoding = 'utf-8')

# drop 'Unnamed:0' columns
nasdaq_final = nasdaq_final.drop(columns = 'Unnamed: 0')
fin_stmts = fin_stmts.drop(columns = 'Unnamed: 0')

display(nasdaq_final)
display(fin_stmts)
```

```python
# merge nasdaq_final and fin_stmts
final_df = pd.merge(nasdaq_final, fin_stmts, how = 'inner', on = 'Ticker')
display(final_df)
```

```python
# reformat the columns' data types for analysis
import numpy as np

final_df['Last Sale'] = final_df['Last Sale'].str.replace('$', '', regex = False)
final_df['Last Sale'] = final_df['Last Sale'].replace(',', '',regex = True).astype(float)

final_df['Percent Change'] = final_df['Percent Change'].str.replace('%', '', regex = False)
final_df['Percent Change'] = final_df['Percent Change'].str.replace('--', '0', regex = False).astype(float)

final_df['Net Change'] = final_df['Net Change'].str.replace('UNCH', '0', regex = False).astype(float)

final_df['Market Cap'] = final_df['Market Cap'].str.replace(',', '', regex = True)
final_df['Market Cap'] = final_df['Market Cap'].replace('NA', np.nan)
final_df['Market Cap'] = final_df['Market Cap'].astype(float)
 
final_df['Share Issued'] = final_df['Share Issued'].astype(float)
final_df['Total Assets'] = final_df['Total Assets'].astype(float) 
final_df['Net PPE'] = final_df['Net PPE'].astype(float) 
final_df['Accounts Receivable'] = final_df['Accounts Receivable'].astype(float) 
final_df['Cash And Cash Equivalents'] = final_df['Cash And Cash Equivalents'].astype(float) 
final_df['Total Debt'] = final_df['Total Debt'].astype(float) 
final_df['Accounts Payable'] = final_df['Accounts Payable'].astype(float) 
final_df['Stockholders Equity'] = final_df['Stockholders Equity'].astype(float) 
final_df['Retained Earnings'] = final_df['Retained Earnings'].astype(float) 
final_df['Working Capital'] = final_df['Working Capital'].astype(float) 
final_df['Total Revenue'] = final_df['Total Revenue'].astype(float) 
final_df['Operating Income'] = final_df['Operating Income'].astype(float)
final_df['EBITDA'] = final_df['EBITDA'].astype(float) 
final_df['Net Income'] = final_df['Net Income'].astype(float) 
final_df['Capital Expenditure'] = final_df['Capital Expenditure'].astype(float) 
final_df['Free Cash Flow'] = final_df['Free Cash Flow'].astype(float) 

# change the df settings to not display values in a scientific notation format
pd.options.display.float_format = '{:.2f}'.format

display(final_df)
```
