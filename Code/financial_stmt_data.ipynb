```python
# import required packages
import pandas as pd
import yfinance as yf
```

```python
# import nasdaq_final.csv
nasdaq_df = pd.read_csv('nasdaq_final.csv', encoding = 'utf-8')
display(nasdaq_df)
```

```python
# randomly select 500 stocks from the nasdaq_df to download Yahoo Finance data on
nasdaq_500 = nasdaq_df.sample(n = 750, random_state = 42)
display(nasdaq_500)
```

```python
# make a list of all the ticker symbols for the 500 randomly-sampled nasdaq stocks
ticker_list = nasdaq_500['Ticker'].unique()
print(ticker_list)
```

```python
# download all the stock information for ticker_list
for ticker in ticker_list:
    stock_information = yf.download(ticker, '2024-04-26')
```

```python
# balance sheet data for each ticker
balance_sheets = []
for ticker in ticker_list:
    tkr = yf.Ticker(ticker)
    balance_sheet = tkr.quarterly_balance_sheet.transpose()
    balance_sheet.insert(0, 'Ticker', tkr.ticker)
    balance_sheets.append(balance_sheet.head(1))
    
final_bs = pd.concat(balance_sheets, ignore_index = True)

display(final_bs)
```

```python
# only include relevant columns
final_bs = final_bs[['Ticker', 'Share Issued', 'Total Assets',
                    'Net PPE', 'Inventory', 'Accounts Receivable',
                    'Cash And Cash Equivalents', 'Total Debt', 'Long Term Debt',
                    'Current Debt', 'Accounts Payable', 'Stockholders Equity',
                    'Retained Earnings', 'Working Capital']]

display(final_bs)
```

```python
# income statement data for each ticker
income_statments = []
for ticker in ticker_list:
    tkr = yf.Ticker(ticker)
    income_statment = tkr.quarterly_financials.transpose()
    income_statment.insert(0, "Ticker", tkr.ticker)
    income_statments.append(income_statment.head(1))

final_is = pd.concat(income_statments, ignore_index=True)

display(final_is)
```

```python
# only include relevant columns
final_is = final_is[[
    'Ticker',                              
    'Total Revenue',
    'Gross Profit',  
    'Operating Income',
    'EBITDA',
    'Net Income',
    'Diluted EPS',
    'Diluted Average Shares',
    'Interest Expense',
]]
display (final_is)
```

```python
# cash flow data for each ticker
cash_flows = []
for ticker in ticker_list:
    tkr = yf.Ticker(ticker)
    cash_flow = tkr.quarterly_cash_flow.transpose()
    cash_flow.insert(0, "Ticker", tkr.ticker)
    cash_flows.append(cash_flow.head(1))

final_cf = pd.concat(cash_flows, ignore_index=True)

display(final_cf)
```

```python
# only include relevant columns
final_cf = final_cf[['Ticker', 'Capital Expenditure',
                    'Free Cash Flow']]
display(final_cf)
```

```python
# merge the balance sheet and income statement dataframe
yf_df_1 = pd.merge(final_bs, final_is, how='inner', on='Ticker')
display(yf_df_1)
```

```python
# merge yf_df_1 with the cash flow statement dataframe
fin_stmts = pd.merge(yf_df_1, final_cf, how = 'inner', on = 'Ticker')
display(fin_stmts)
```

```python
# check for missing values
fin_stmts.isna().sum()
```

```python
# drop columns with over 100 of the observations missing
fin_stmts = fin_stmts.drop(columns = ['Inventory', 'Long Term Debt', 'Current Debt', 'Diluted EPS', 'Gross Profit', 'Diluted Average Shares', 'Interest Expense'])
fin_stmts.isna().sum()
```

```python
# save fin_stmts as a .csv (fin_stmts.csv)
fin_stmts.to_csv('fin_stmts.csv', encoding = 'utf-8')
```
